```python
st =['AZ', 'CA', 'CO', 'DC', 'FL', 'IL', 'IN', 'KS', 'KY', 'LA', 'MI',
     'MO', 'NJ', 'NY', 'NC', 'OK', 'PA', 'TX', 'VA', 'WA', 'WI', 'MA',
     'NV', 'NH', 'TN', 'SC', 'CT', 'IA', 'ME', 'MD', 'AR', 'OG', 'WY',
     'ND', 'ID', 'OH', 'GA', 'DE', 'HI', 'MN', 'NM', 'RI', 'SD', 'UT']
i = 2000
lis=[]
while i<2013:
    df2= df[df["year"]==i]
    for x in st:
        df3= df2[df2["State"]==x]
        dic = {"year": i,"O3mean":df2["O3 Mean"].mean(),
               "COmean":df2["CO Mean"].mean(), "SO2mean":df2["SO2 Mean"].mean(), "NO2mean":df2["NO2 Mean"].mean(),
              "State":x}
        lis.append(dic)
    i+=1
df = pd.DataFrame.from_dict(lis)
df
```
